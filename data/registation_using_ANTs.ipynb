```python
import nibabel as nib
import matplotlib.pyplot as plt
import ants
import numpy as np
import os
```

```python
image_type ="pet"
resolution = (128,128,64)
input_folder = "/home/jovyan/work/head-neck-ordered"
output_folder = os.path.join("/home/jovyan/work/head-neck-reg-small",image_type)
folder_list = [os.path.join(input_folder,f) for f in os.listdir(input_folder) if os.path.isdir(os.path.join(input_folder,f))]
print(len(folder_list))
```

```python
fixed_folder = os.path.join(input_folder,"HN-CHUM-007",image_type)
fixed = ants.image_read(os.path.join(fixed_folder,os.listdir(fixed_folder)[0]))
fixed = ants.resample_image(fixed, resolution, 1, 0)
plt.imshow(fixed[:,:,5])
```

```python
i=0
for folder in folder_list:
    folder = os.path.join(folder,image_type) 
    if(len(os.listdir(folder))==0):
        print("file: " +os.path.basename(os.path.split(folder)[0]) + " not found" )
        continue
    filename = os.path.join(folder,os.listdir(folder)[0])
    moving = ants.image_read(filename)
    moving = ants.resample_image(moving, resolution, 1, 0)
    
    mytx = ants.registration(fixed=fixed , moving=moving, type_of_transform='Translation' )
    warped_moving = mytx['warpedmovout']
    warped_movinf = nib.Nifti1Image(warped_moving.numpy(), np.eye(4))
    
    save_location = os.path.join(output_folder,os.path.basename(os.path.split(folder)[0])+".nii.gz")
    nib.save(warped_moving, save_location)
    print("saved at :  " + save_location)
    i+=1
print("number of images created: " + str(i))
```
